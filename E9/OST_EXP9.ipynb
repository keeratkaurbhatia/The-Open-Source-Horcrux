```python
# Temperature Converter: Write a program that converts Celsius to Fahrenheit or Fahrenheit to Celsius, depending on user input.

#inputs
y = input('enter celsius or fahrenheit-').lower()

while y != "celsius" and y != "fahrenheit": #otherwise, retry
  print('invalid input')
  y = input('enter celsius or fahrenheit-').lower()


x = float(input('enter temp value-'))

if y == "celsius":
  a = ((x*9)/5) + 32 # formula conversion
  print(a, 'degree fahrenheit')
elif y == "fahrenheit":
  a = ((x-32)*5)/9 # conversion
  print(a, 'degree celsius')


```

```python
"""
Leap Year or Not: Write a program that checks if a given year is a leap year or not.
Google for the details on how to figure out if the given number is a leap year or not. It is more complicated than simply checking for a multiple of 4.
"""
# RULE- DIVISIBLE BY 4 , EXCEPT END OF CENTURY YEARS, WHICH NEEDS TO BE DIVISIBLE BY 400

x = int(input('enter the year-'))

if x % 4 == 0: # if divisible by 4,
  if x % 100 == 0: # check if its the end of century
    if x % 400 == 0: # if yes, then checking whether its leap year or not
      print('leap year')
    elif x % 400 != 0:
      print('not a leap year')
  else: # if not end of century but div by 4, then leap year.
    print('leap year')

else:
  print('not a leap year')
```

```python
# Write a program that takes as inputs list of lists that depicts square matrices A and B and then gives us the sum.

def sum_matrices(A, B):
  result = []
  for i in range(len(A)):
    row = []
    for j in range(len(A[0])):
      row.append(A[i][j] + B[i][j])
    result.append(row)
  return result

# Example usage:
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]
C = sum_matrices(A, B)
print(C)  # Output: [[6, 8], [10, 12]]
```
