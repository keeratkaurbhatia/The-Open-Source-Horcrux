```python

import pandas as pd
df = pd.read_csv('first_year_record.csv')

# viewing the first few rows
print(df.head())
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# barplot
plt.figure(figsize=(12, 6))
sns.barplot(x='Subject', y='Percentage', data=df, palette='Set2')
plt.title('Percentage per Subject')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 100)
plt.show()

```

```python


#lineplot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Subject', y='Percentage', data=df, marker='o', markersize=10, linewidth=3, color='#FF5733')
plt.title('Trend of Marks Across Subjects')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 100)
plt.fill_between(range(len(df)), df['Percentage'], alpha=0.2, color='#FF5733')

plt.show()

```

```python


# sorted horizontal barplot
sorted_df = df.sort_values(by='Percentage', ascending=True)
plt.figure(figsize=(12, 6))
sns.barplot(x='Percentage', y='Subject', data=sorted_df, palette='flare')
plt.title('Subjects Sorted by Marks')
plt.xlim(0, 100)
plt.show()

```

```python


# pie chart
plt.figure(figsize=(10,10))
plt.pie(df['Percentage'], labels=df['Subject'], autopct='%1.1f%%', startangle=90,
        colors=sns.color_palette('pastel'))
plt.title('Result Contribution by Subject (Percentage)')
plt.tight_layout()
plt.show()
```
